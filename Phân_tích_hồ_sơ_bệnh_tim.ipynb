```python
import pandas as pd
from pymongo import MongoClient

# tạo database
db = client['heart_disease_db']

# Xóa collection cũ nếu có
db.patients.drop()
db.lifestyle.drop()
db.health_records.drop()

# Tạo 3 documents list
patients_docs = []
lifestyle_docs = []
health_records_docs = []

for idx, row in df_2020.iterrows():
    pid = str(idx)
    patients_docs.append({
        "_id": pid,
        "sex": row["Sex"],
        "age_category": row["AgeCategory"],
        "race": row["Race"],
         "bmi": float(row["BMI"])
    })
    lifestyle_docs.append({
        "patient_id": pid,
        "smoking": row["Smoking"],
        "alcohol_drinking": row["AlcoholDrinking"],
        "physical_activity": row["PhysicalActivity"],
        "sleep_time": int(row["SleepTime"]),
        "general_health": row["GenHealth"],
    })
    health_records_docs.append({
        "patient_id": pid,
        "heart_disease": row["HeartDisease"],
        "stroke": row["Stroke"],
        "diabetic": row["Diabetic"],
        "diffWalking": row["DiffWalking"],
        "physical_health": int(row["PhysicalHealth"]),
        "mental_health": int(row["MentalHealth"]),
        "asthma": row["Asthma"],
        "kidney_disease": row["KidneyDisease"],
        "skin_cancer": row["SkinCancer"]
    })

# Insert vào MongoDB
db.patients.insert_many(patients_docs)
db.lifestyle.insert_many(lifestyle_docs)
db.health_records.insert_many(health_records_docs)

print("✅ Import thành công!")
```

```python
pipeline = [
    # B1: Ghép health_records với patients để lấy chỉ số BMI
    {"$lookup": {
        "from": "patients",
        "localField": "patient_id",
        "foreignField": "_id",
        "as": "patient_info"
    }},
    {"$unwind": "$patient_info"},

    # B2: Tạo trường mới để phân nhóm BMI
    {"$addFields": {
        "bmi_group": {
            "$switch": {
                "branches": [
                    {"case": {"$lt": ["$patient_info.bmi", 18.5]}, "then": "<18.5"},
                    {"case": {"$and": [{"$gte": ["$patient_info.bmi", 18.5]}, {"$lt": ["$patient_info.bmi", 25]}]}, "then": "18.5–24.9"},
                    {"case": {"$and": [{"$gte": ["$patient_info.bmi", 25]}, {"$lt": ["$patient_info.bmi", 30]}]}, "then": "25–29.9"},
                    {"case": {"$and": [{"$gte": ["$patient_info.bmi", 30]}, {"$lt": ["$patient_info.bmi", 35]}]}, "then": "30–34.9"},
                    {"case": {"$and": [{"$gte": ["$patient_info.bmi", 35]}, {"$lt": ["$patient_info.bmi", 40]}]}, "then": "35–39.9"},
                ],
                "default": ">=40"
            }
        }
    }},

    # B3: Gom nhóm theo bmi_group, tính tổng số và số người bệnh tim
    {"$group": {
        "_id": "$bmi_group",
        "total": {"$sum": 1},
        "heart_disease_count": {
            "$sum": {"$cond": [{"$eq": ["$heart_disease", True]}, 1, 0]}
        }
    }},

    # B4: Tính tỷ lệ mắc bệnh tim trong mỗi nhóm
    {"$project": {
        "bmi_group": "$_id",
        "_id": 0,
        "total": 1,
        "heart_disease_count": 1,
        "heart_disease_rate": {
            "$multiply": [
                {"$divide": ["$heart_disease_count", "$total"]},
                100
            ]
        }
    }},

    # B5: Sắp xếp theo nhóm BMI (tuỳ chọn)
    {"$sort": {"bmi_group": 1}}
]

results = list(db.health_records.aggregate(pipeline))
for r in results:
    print(f"BMI nhóm {r['bmi_group']}:")
    print(f"  Tổng số người: {r['total']}")
    print(f"  Người có bệnh tim: {r['heart_disease_count']}")
    print(f"  Tỷ lệ mắc bệnh tim: {r['heart_disease_rate']:.2f}%\n")
```

```python
pipeline = [
    # Bước 1: Nối thông tin lifestyle (để lấy hành vi hút thuốc)
    {
        "$lookup": {
            "from": "lifestyle",
            "localField": "patient_id",
            "foreignField": "patient_id",
            "as": "lifestyle"
        }
    },
    {"$unwind": "$lifestyle"},

    # Bước 2: Nhóm theo hành vi hút thuốc, tính tổng số người và số người mắc bệnh tim
    {
        "$group": {
            "_id": "$lifestyle.smoking",
            "total": { "$sum": 1 },
            "heart_disease_count": {
                "$sum": { "$cond": [{ "$eq": ["$heart_disease", True] }, 1, 0] }
            }
        }
    },

    # Bước 3: Tính tỷ lệ mắc bệnh tim trong từng nhóm
    {
        "$project": {
            "_id": 0,
            "smoking": "$_id",
            "total": 1,
            "heart_disease_count": 1,
            "heart_disease_rate": {
                "$multiply": [
                    { "$divide": ["$heart_disease_count", "$total"] },
                    100
                ]
            }
        }
    }
]

results = list(db.health_records.aggregate(pipeline))

for r in results:
    status = "Hút thuốc" if r['smoking'] else "Không hút thuốc"
    print(f"{status}:")
    print(f"  Tổng số người: {r['total']}")
    print(f"  Số người mắc bệnh tim: {r['heart_disease_count']}")
    print(f"  Tỷ lệ mắc bệnh tim: {r['heart_disease_rate']:.2f}%\n")
```

```python
pipeline = [
    # Ghép bảng lifestyle vào health_records
    {
        "$lookup": {
            "from": "lifestyle",
            "localField": "patient_id",
            "foreignField": "patient_id",
            "as": "lifestyle"
        }
    },
    # Tách mảng lifestyle (chỉ lấy 1 lifestyle cho mỗi bệnh nhân)
    {"$unwind": "$lifestyle"},

    # Lọc bỏ dữ liệu thiếu hoặc không hợp lệ
    {
        "$match": {
            "lifestyle.sleep_time": {"$exists": True, "$ne": None},
            "heart_disease": {"$in": [True, False]}
        }
    },

    # Gom nhóm theo trạng thái bệnh tim
    {
        "$group": {
            "_id": "$heart_disease",
            "avg_sleep": {"$avg": "$lifestyle.sleep_time"},
            "count": {"$sum": 1}
        }
    },

    # Hiển thị rõ ràng
    {
        "$project": {
            "_id": 0,
            "heart_disease": "$_id",
            "avg_sleep_time": {"$round": ["$avg_sleep", 2]},
            "number_of_people": "$count"
        }
    }
]

# Thực thi và in kết quả
results = list(db.health_records.aggregate(pipeline))
for r in results:
    status = "CÓ bệnh tim" if r["heart_disease"] else "KHÔNG có bệnh tim"
    print(f"{status}: Trung bình ngủ {r['avg_sleep_time']} giờ/ngày (n = {r['number_of_people']})")
```

```python
# Mối quan hệ giữa giờ ngủ và bệnh tim	Cột (giờ ngủ vs % mắc bệnh)	Thể hiện phân tích nghiêm túc
pipeline = [
    {
        "$lookup": {
            "from": "lifestyle",
            "localField": "patient_id",
            "foreignField": "patient_id",
            "as": "lifestyle"
        }
    },
    {"$unwind": "$lifestyle"},
    {
        "$match": {
            "lifestyle.sleep_time": {"$gte": 3, "$lte": 12},  # Giới hạn giờ ngủ hợp lý
            "heart_disease": {"$in": [True, False]}
        }
    },
    {
        "$group": {
            "_id": {
                "sleep_time": "$lifestyle.sleep_time",
                "heart_disease": "$heart_disease"
            },
            "count": {"$sum": 1}
        }
    },
    {
        "$group": {
            "_id": "$_id.sleep_time",
            "counts": {
                "$push": {
                    "heart_disease": "$_id.heart_disease",
                    "count": "$count"
                }
            },
            "total": {"$sum": "$count"}
        }
    },
    {
        "$project": {
            "_id": 0,
            "sleep_time": "$_id",
            "rate_heart_disease": {
                "$multiply": [
                    {
                        "$divide": [
                            {
                                "$sum": {
                                    "$map": {
                                        "input": "$counts",
                                        "as": "c",
                                        "in": {
                                            "$cond": [{"$eq": ["$$c.heart_disease", True]}, "$$c.count", 0]
                                        }
                                    }
                                }
                            },
                            "$total"
                        ]
                    },
                    100
                ]
            }
        }
    },
    {"$sort": {"sleep_time": 1}}
]

results = list(db.health_records.aggregate(pipeline))
# hiện nhóm kết quả
for r in results:
    print(f"Giờ ngủ: {r['sleep_time']} - Tỷ lệ mắc bệnh tim: {r['rate_heart_disease']:.2f}%")
    print()

```

```python
pipeline = [
    # B1: Nối health_records với patients
    {"$lookup": {
        "from": "patients",
        "localField": "patient_id",
        "foreignField": "_id",
        "as": "patient_info"
    }},
    {"$unwind": "$patient_info"},

    # B2: Nhóm theo age_category và sex
    {"$group": {
        "_id": {
            "age": "$patient_info.age_category",
            "sex": "$patient_info.sex"
        },
        "total": {"$sum": 1},
        "heart_disease_count": {
            "$sum": {
                "$cond": [{"$eq": ["$heart_disease", True]}, 1, 0]
            }
        }
    }},

    # B3: Tính tỷ lệ bệnh tim
    {"$project": {
        "_id": 0,
        "age": "$_id.age",
        "sex": "$_id.sex",
        "total": 1,
        "heart_disease_count": 1,
        "heart_disease_rate": {
            "$round": [
                {"$multiply": [
                    {"$divide": ["$heart_disease_count", "$total"]}, 100
                ]},
                2  # Làm tròn 2 chữ số
            ]
        }
    }},

    # B4: Sắp xếp cho dễ xem (tuổi từ nhỏ đến lớn)
    {"$sort": {"age": 1, "sex": 1}}
]

results = list(db.health_records.aggregate(pipeline))
for r in results:
    print(f"{r['age']} - {r['sex']}: {r['heart_disease_rate']}% ({r['heart_disease_count']}/{r['total']})")
```

```python
pipeline = [

    # Bước 1: Chỉ giữ lại những bản ghi có dữ liệu hợp lệ
    {
        "$match": {
            "heart_disease": { "$in": [True, False] },
            "stroke": { "$in": [True, False] }
        }
    },

    # Bước 2: Nhóm theo tình trạng bệnh tim (True/False)
    {
        "$group": {
            "_id": "$heart_disease",  # Nhóm theo: có bệnh tim hay không
            "total": { "$sum": 1 },   # Tổng số người trong mỗi nhóm
            "stroke_count": {
                "$sum": {
                    "$cond": [{ "$eq": ["$stroke", True] }, 1, 0]
                }
            }
        }
    },

    # Bước 3: Tính tỷ lệ % bị đột quỵ trong mỗi nhóm
    {
        "$project": {
            "_id": 0,
            "heart_disease": "$_id",
            "total": 1,
            "stroke_count": 1,
            "stroke_rate": {
                "$round": [
                    { "$multiply": [
                        { "$divide": ["$stroke_count", "$total"] }, 100
                    ]}, 2
                ]
            }
        }
    }
]

results = list(db.health_records.aggregate(pipeline))
labels = []
counts = []

for r in results:
    label = "CÓ bệnh tim" if r["heart_disease"] else "KHÔNG có bệnh tim"
    labels.append(label)
    counts.append(r["stroke_count"])  # Số người bị đột quỵ trong mỗi nhóm
    print(f"{label}:")
    print(f"  Tổng số người: {r['total']}")
    print(f"  Số người từng bị đột quỵ: {r['stroke_count']}")
    print(f"  Tỷ lệ bị đột quỵ: {r['stroke_rate']}%\n")

```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title("Tỷ lệ người bị bệnh tim và đột quỵ")
plt.axis('equal')  # Vẽ hình tròn đều
plt.tight_layout()
plt.show()
```
